# Altair test
Altair has been buggy with jupyter lab since the 3.0 release. I specifically ran the following:

    pip install jupyterlab==1.0.0a3
    
I am testing to make sure this works. Altair defaults to over 3.0 now, so you supposedly only need to specifically pip install a version of jupyterlab over 1.0

```python
import pandas as pd
import altair as alt
import numpy as np
```

```python
df = pd.DataFrame(columns=["a","b"])
df.a = pd.Series([1,2,3,4,5])
df.b = pd.Series([1,2,3,4,5])
```

```python
chart = alt.Chart(df).mark_line()
```

```python
chart.encode(
x="a",
y="b")
```

Success!
